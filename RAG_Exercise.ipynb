Now that you are familiar with the structure of the dataset, perform the query. For this exercise you dont need to make a dynamic query, but a simple search should be enough.

Please search text related to the Declaration of the Rights of Man and the Citizen, and choose one string.

```python
article_text = ''
```

## We are providing you with the prompt in this case:

Tell me about the "Tell me about the Declaration of the rights of man and the citizen"

Here we are also combining it with the article text.

```python
custom_prompt = "Tell me about the Declaration of the rights of man and the citizen"
combined_prompt = custom_prompt + article_text
```

The following code will process your inputs through the gpt2 model. It will take a while to run, but after you are done, that's it you've ran your first very basic RAG model!

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')

inputs = tokenizer.encode(combined_prompt, return_tensors='pt')
outputs = model.generate(inputs, max_length=1024, num_return_sequences=1)
```

```python
decoded_output = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(decoded_output)
```
